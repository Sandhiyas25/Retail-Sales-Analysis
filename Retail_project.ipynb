```python
df['order_date'] = pd.to_datetime(df['order_date'], format='%d-%m-%Y', errors='coerce')
```

```python
df['Recency']=(current_date-df['order_date']).dt.days
print(df)
```

```python
orderwisecount=orders.groupby('customer_id')['order_id'].agg(Frequency='count')
```

```python
orderwisecount
```

```python
merge_data=pd.merge(orders,order_items,on='order_id',how='inner')
```

```python
merge_data
```

```python
print(merge_data.head())
```

```python
merge_data['monetary']=merge_data['quantity']*merge_data['list_price']
```

```python
print(merge_data.head())
```

```python
orderwisepurchase=merge_data.groupby('customer_id')['monetary'].agg(monetary='sum')
print(orderwisepurchase)
```

```python
rfm=pd.merge(pd.merge(df,orderwisecount,on='customer_id'),orderwisepurchase,on='customer_id')
```

```python
rfm
```

```python
rfm.drop('order_date',axis=1,inplace=True)
```

```python
rfm
```

```python
rfm.to_sql('customer_segments', con=engine, if_exists='replace', index=False)
```
